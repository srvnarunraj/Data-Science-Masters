# Q1. Explain GET and POST methods.

>* **Get :**  
	* GET requests are typically used to retrieve data, and they can be bookmarked and cached by web browsers. 
	* GET request params can be seen in url
>* **POST :**
	* POST used to submit data to a server, such as the contents of a form or a database query, and it sends data in the request body. 
    * POST requests are typically used to create, update, or delete data, and they cannot be cached or bookmarked.
    * Data isnt visible in the URL

# Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
@app.route('/psman',methods=['GET','POST'])
def math_ops():
    if request.method == 'POST':
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1 + num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        elif ops == 'subtract':
            r = num1 - num2
            result = "The difference between " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        elif ops == 'multiply':
            r = num1 * num2
            result = "The product of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        elif ops == 'divide':
            r = num1 / num2
            result = "The quotient of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        else:
            result = "Invalid operation"
        return jsonify(result)
```
