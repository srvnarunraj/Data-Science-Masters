* The Exception class is the base class for all exceptions in Python, and it provides a common interface for handling exceptions.
* When you create a custom exception, it is recommended to inherit from the Exception class so that you can take help from already defined methods. 
* custom exceptions to be caught by catch blocks that are handling exceptions of the Exception type, which is the most general type of exception.

## Q2. Write a python program to print Python Exception Hierarchy.

```python
def printExceptions(exception, level=0):
    print(" " * level + exception.__name__)
    for i in exception.__subclasses__():
        printExceptions(i, level + 4)

printExceptions(Exception)
```

## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.


Errors in Dictionary / Lists are handled by Lookupclass. The KeyError and IndexError are two common types of LookupError exceptions.
* a. KeyError: Raised when a dictionary key is not found.
>*Example: d = {'a': 1} and d['b']

* b. IndexError: Raised when an index is not within the bounds of a list.
>* Example: l = [1, 2, 3] and l[3]
