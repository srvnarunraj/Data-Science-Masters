### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Set the sample mean and standard deviation
sample_mean = 50
sample_std = 5
sample_size = 100

# Set the significance level and calculate the z-score
significance_level = 0.05
z_score = norm.ppf(1 - (significance_level / 2))

# Calculate the margin of error
margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print(f"The 95% confidence interval is [{lower_bound:.2f}, {upper_bound:.2f}].")
```

### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

# Define the observed frequencies
observed_freq = [10, 20, 30, 15, 5, 20]

# Define the expected frequencies
expected_freq = [0.2 * sum(observed_freq)] * 6
expected_freq[0] = 0.2 * sum(observed_freq) # blue
expected_freq[1] = 0.2 * sum(observed_freq) # orange
expected_freq[2] = 0.2 * sum(observed_freq) # green
expected_freq[3] = 0.1 * sum(observed_freq) # yellow
expected_freq[4] = 0.1 * sum(observed_freq) # red
expected_freq[5] = 0.2 * sum(observed_freq) # brown

# Perform the chi-square test
statistic, p_value = chisquare(observed_freq, expected_freq)

# Set the significance level
significance_level = 0.05

# Print the results
print("Chi-square statistic:", statistic)
print("P-value:", p_value)
if p_value < significance_level:
    print("Reject null hypothesis - the distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject null hypothesis - the distribution of colors matches the expected distribution.")
```

### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked.
Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import math
p = 60/500
z = 1.96 # 95% confidence level
se = math.sqrt((p*(1-p))/500)
lower = p - z*se
upper = p + z*se
print("95% confidence interval:", (lower, upper))
```

### Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import scipy.stats as stats
import math
n = 1 # sample size
mean = 75
std_dev = 12
z = stats.norm.ppf(0.95) # 90% confidence level
se = std_dev / math.sqrt(n)
lower = mean - z*se
upper = mean + z*se
print("90% confidence interval:", (lower, upper))
```

### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 10
x = np.linspace(0, 30, 5000)
plt.plot(x, stats.chi2.pdf(x, df))
plt.fill_between(x[x>=15], stats.chi2.pdf(x[x>=15], df), color='red')
plt.title("Chi-Square Distribution with 10 Degrees of Freedom")
plt.xlabel("Chi-Square Statistic")
plt.ylabel("Density")
plt.show()
```

### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520  preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who  prefer Coke.

### Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails.
Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

```python
import scipy.stats as stats
import numpy as np

observed_freq = [55, 45] # number of tails and heads observed
expected_freq = [50, 50] # number of tails and heads expected for a fair coin
# Calculate the chi-square test statistic and p-value
chi_sq_stat, p_value = stats.chisquare(observed_freq, expected_freq)

# Calculate the critical value at the 0.05 significance level with 1 degree of freedom
crit_val = stats.chi2.ppf(0.95, 1)

# Compare the chi-square test statistic to the critical value and the p-value to the significance level
if chi_sq_stat > crit_val:
    print("Reject null hypothesis: The coin is biased towards tails.")
else:
    print("Fail to reject null hypothesis: The coin is not biased towards tails.")
    
if p_value < 0.05:
    print("Reject null hypothesis: The coin is biased towards tails.")
else:
    print("Fail to reject null hypothesis: The coin is not biased towards tails.")
```

### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation.
The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

```python
import numpy as np
from scipy import stats

# Define the sample size, sample mean, and sample standard deviation
n = 30
x_bar = 72
s = 10

# Define the null hypothesis mean and the significance level
mu0 = 70
alpha = 0.05

# Calculate the t-test statistic and the p-value
t_stat, p_value = stats.ttest_1samp([x_bar], mu0, alternative="two-sided")
print(t_stat, p_value)

# Calculate the critical value at the 0.05 significance level with 29 degrees of freedom
crit_val = stats.t.ppf(1 - alpha/2, n-1)

# Compare the t-test statistic to the critical value and the p-value to the significance level
if np.abs(t_stat) > crit_val:
    print("Reject null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject null hypothesis: The population mean is not significantly different from 70.")
    
if p_value < alpha:
    print("Reject null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject null hypothesis: The population mean is not significantly different from 70.")
```
