```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


* Seaborn is a Python data visualization library that is built on top of Matplotlib. It provides a high-level interface  for creating informative and attractive statistical graphics

1. **Scatter Plot:** plotted along two axes, revealing any correlation present
2. **Line Plot:** A line plot is a type of graph that displays information as a series of data points connected by straight line segments.
3. **Bar Plot:** A bar plot is a graph that represents categorical data with rectangular bar
4. **Histogram:** A histogram is a graph showing the frequency distribution of a set of continuous data. 
5. **Heatmap:** A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 
<pre>
Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
</pre>

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',style='region',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',style='event',data=fmri)
```


### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. 
<pre>
Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
</pre>

```python
tit = sns.load_dataset('titanic')
sns.boxplot(x='pclass',y='age',data=tit)
plt.show()
sns.boxplot(x='pclass',y='fare',data=tit)
plt.show()
```


### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.


```python
dia = sns.load_dataset('diamonds')
```

```python
sns.displot(x='price',hue='cut',data=dia)
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset('iris')
sns.pairplot(data=iris,hue='species')
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flight = sns.load_dataset('flights')
flight
```

```python
flightmatrix = flight.pivot('year','month','passengers')
```

```python
sns.heatmap(flightmatrix,annot=True,fmt='d')
```
